```python
online_orders = pd.DataFrame(cmd_internet.groupby('ORDER_DATE').CLIENT_NUMBER.nunique())
online_orders = online_orders.reset_index()
online_orders.head()
```

```python
pd.to_datetime(online_orders['ORDER_DATE'])
online_orders.head()
```

```python
# Considering only the period from 2017-08-08 to 2018-03-31

adjusted_nb_of_orders = online_orders.loc[(online_orders['ORDER_DATE'] >= '2017-08-08') & (online_orders['ORDER_DATE'] <= '2018-03-31')]
adjusted_nb_of_orders.head()
```

```python
total_nb_purchase = adjusted_nb_of_orders['CLIENT_NUMBER'].sum()
total_nb_purchase
```

# LOGS

```python
# We merge all log files, from 08-08-2017 to 31-03-2018

import glob 
log_path = glob.glob('Data/LOG*.csv')
logs = pd.DataFrame()
log_list = []
for log_file in log_path:
   df = pd.read_csv(log_file, index_col=None, delimiter=';')
   log_list.append(df)
logs = pd.concat(log_list)
```

```python
logs_table = logs
```

```python
logs_table.loc[:, 'SESSION_DATE'] = pd.to_datetime(logs_table['SESSION_START_DATE']).dt.date
logs_table
```

```python
logs_table.to_csv('logs.csv')
```

```python
# Session ids are reset everyday. To count the number of visits over the 7 %morenths we observe, we compute the sum of the number of unique session_id per day
nb_of_id_sessions = pd.DataFrame(logs_table.groupby('SESSION_DATE').ID_SESSION.nunique())
```

```python
nb_of_id_sessions.head()
```

```python
total_visits = nb_of_id_sessions['ID_SESSION'].sum()
total_visits
```

## The total number of visits from 08-08-2017 to 31-03-2018 is 7 241 180.

```python
# Conversion rate = percentage of visitors who make an order
```

```python
rate = total_nb_purchase/total_visits
rate
```

# From 08-08-2017 to 31-03-2018, the conversion rate is about 2,6%.

```python
logs['PAGES'].value_counts()
```

```python
panier = pd.DataFrame(logs_table[logs_table['PAGES'] == 'ajout au panier'])
panier
```

```python
len(panier)
```

```python
# Computing the number of people (unique session_id) who put items in their cart per day

panier_per_day = pd.DataFrame(panier.groupby('SESSION_DATE').ID_SESSION.nunique())
panier_per_day.head()
```

```python
total_panier = panier_per_day['ID_SESSION'].sum()
total_panier
```

CART CONVERSION RATE

```python
cart_conversion = total_nb_purchase/total_panier
cart_conversion
```

FROM VISIT TO CART

```python
add_to_cart = total_panier/total_visits
add_to_cart
```

```python
add_to_cart * cart_conversion
```
