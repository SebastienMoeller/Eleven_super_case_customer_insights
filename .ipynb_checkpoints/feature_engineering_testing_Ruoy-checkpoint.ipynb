## for the CMD file:

trim it by filtering by date

filter by order channel

then create new column: client # + order date

keep lines with unique values in this column

```python
#trimming the CMD file as we only wanna keep details of orders made from 201708 onwards and of orders made v
def desired_group_2(row):
    if row['ORDER_CHANNEL'] == 'Internet':
        if row['ORDER_DATE'][:4] == '2017':
            if int(row['ORDER_DATE'][5:7]) >= 8:
                return(True)
        if row['ORDER_DATE'][:4] == '2018':
            return(True)



CMD_slim = CMD[[True if desired_group_2(row) else False for i, row in tqdm(CMD.iterrows())]]
```

```python
len(CMD), len(CMD_slim)
```

```python
CMD_slim = CMD_slim.reset_index(drop = True)
```

```python
CMD_slim.to_csv(data_path+'CMD_slim.csv')
```

## for the CMD file: Create new col: client # + order date

```python
CMD_slim['identifiant'] = [str(CMD_slim.CLIENT_NUMBER[i]) + str(CMD_slim.ORDER_DATE[i]) for i in range(len(CMD_slim))]
```

we remove lines of clients making multiple orders on the same day

```python
CMD_idx_temp = [k for k,v in Counter(CMD_slim.identifiant).items() if v == 1]
```

```python
len(CMD_idx_temp), len(CMD_slim)
```

```python
CMD_slim.index = CMD_slim.identifiant
CMD_slim = CMD_slim.loc[CMD_idx_temp]
len(set(CMD_slim.identifiant)), len(CMD_slim)
```
